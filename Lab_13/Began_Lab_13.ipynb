```python
# Importing required libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
# load dataset
data = pd.read_csv("Advertising.csv")
print(data.head())
```

```python
# Preprocessing data
data.drop(['Unnamed: 0'], axis=1)
```

```python
#scatter plot.
plt.figure(figsize=(16, 8))
plt.scatter(data['TV'],data['Sales'],c='black')
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
#Feature selection :split dataset
X = data['TV'].values.reshape(-1,1)
y = data['Sales'].values.reshape(-1,1)
print("\n -------------Simple Linear Regression--------------\n \n")
reg = LinearRegression()
reg.fit(X, y)
print("The linear model is: Y = {:.5} + {:.5}X\n\n".format(reg.intercept_[0], reg.coef_[0][0]))
```

```python
predictions = reg.predict(X)
plt.figure(figsize=(16, 8))
plt.scatter(data['TV'],data['Sales'], c='black')
plt.plot(data['TV'],predictions, c='blue',linewidth=2)
plt.xlabel("Money spent on TV ads ($)")
plt.ylabel("Sales ($)")
plt.show()
```

```python
#Assessing the relevancy of the model
X = data['TV']
y = data['Sales']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
#Multiple Linear Regression
print("\n ------------------------------------Multiple Linear Regression------------------------\n \n")
#Feature selection :split dataset
Xs = data.drop(['Sales', 'Unnamed: 0'], axis=1)
y = data['Sales']. values.reshape(-1,1)
reg = LinearRegression()
reg.fit(Xs, y)
print("The linear model is: Y = {:.5} + {:.5}*TV + {:.5}*radio + {:.5}*newspaper\n\n".format(reg.intercept_[0], reg.coef_[0][0], reg.coef_[0][1], reg.coef_[0][2]))
```

```python
#Assessing the relevancy of the model
X = np.column_stack((data['TV'], data['Radio'], data['Newspaper']))
y = data['Sales']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
print("\n -------------------Drop Newspaper-------------\n \n")
#Feature selection :split dataset
Xs = data.drop(['Sales', 'Newspaper', 'Unnamed: 0'], axis=1)
y = data['Sales']. values.reshape(-1,1)
reg = LinearRegression()
reg.fit(Xs, y)
print("The linear model is: Y = {:.5} + {:.5}*TV + {:.5}*radio\n\n".format(reg.intercept_[0], reg.coef_[0][0], reg.coef_[0][1]))
```

```python
#Assessing the relevancy of the model
X = np.column_stack((data['TV'], data['Radio']))
y = data['Sales']
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
```
